```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
#Import data
sales = pd.read_csv("supermarket_sales - Sheet1.csv")
```

```python
#Inspecting a DataFrame
sales.head()
```

```python
sales.describe()
```

```python
# Sort sales by gender and  qty
sales_by_gender = sales.sort_values(["Gender","Quantity"], ascending=[True, True])

print(sales_by_gender.head())
```

```python
# Print the mean of "Quantity"
print(sales["Quantity"].mean())

# Print the median of "Quantity"
print(sales["Quantity"].median())

# Print the maximum "Quantity"
print(sales["Quantity"].max())

# Print the minimum "Quantity"
print(sales["Quantity"].min())
```

```python
# create custom IQR function

def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)

# Update to print IQR and median of temperature_c, fuel_price_usd_per_l, & unemployment
print(sales[["Quantity", "Total", "gross margin percentage"]].agg([iqr, np.median]))
```

```python
# Sort sales_by_gender by "Total"
sales_by_gender = sales_by_gender.sort_values("Date", ascending=True)

# Get the cumulative sum of weekly_sales, add as cum_weekly_sales col

sales_by_gender['cum_weekly_sales'] = sales_by_gender["Total"].cumsum()


# Get the cumulative max of weekly_sales, add as cum_max_sales col
sales_by_gender["cum_max_sales"] = sales_by_gender[ "Total"].cummax()

# See the columns you calculated
print(sales_by_gender[["Date", "Total", "cum_weekly_sales", "cum_max_sales"]])
```
